# Project FourSquare

```python
name = input("Enter Name of type of place which pics you want ?")
sort = input("Choose the sorting method among : NEWEST / POPULAR /RATING / RELEVANCE").upper()
if sort == "" : 
    sort = "NEWEST"
classification = input("Enter what type of IMGs you want among : indoor/outdoor/main").lower()
if classification == "" : 
    classification = "indoor"
# loc = input("Enter loction where you want to store all IMG")



api = 'fsq3eJCWdl4Ptag8PDA3POT7hN0fuKzOJ851av+szhYaZJI='

url = "https://api.foursquare.com/v3/places/search"

headers = {
    "Accept": "application/json",
    "Authorization": api
}
par = {
  	"query": name,
  	"ll": "26.9124,75.7873",
  	"open_now": "true",
  	"sort":"DISTANCE"
}
response = requests.get(url, headers=headers, params=par)

res = response.json()


fsq_ids = []
for i in range(len(res['results'])) :
    fsq_ids.append(res['results'][i]['fsq_id'])
    
headers = {
    "Accept": "application/json",
    "Authorization": api
}
par = {
  	'limit' : 20,
    'sort' : sort,
    'classifications' : classification

}

res = []
for i in fsq_ids: 
    url = f"https://api.foursquare.com/v3/places/{i}/photos"
    response = requests.get(url, headers=headers, params=par)
    if len(response.json()) != 0 :
        res.append(response.json())
        
r = 1
for i in res:
    lnk = i[0]['prefix']+"original"+i[0]['suffix']
    rsp = requests.get(lnk)
    loc=""
    loction = f"{loc}FS_IMG/{name}/"
    os.makedirs(loction,exist_ok=True)
    if rsp.status_code == 200 :
        with open(f"{loction}IMG_{r}.jpg", "wb") as f:
            f.write(rsp.content)
            r = r + 1
            f.close()
print("IMGs Fatched sucessfull ")
        
```
